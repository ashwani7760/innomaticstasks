# Problem 1: Social Media – Post Engagement Analyzer

```python
# function that check post is viral or not based on likes
def viral(l):
    
    # dictionary that contains likes and its status
    likes ={}
    
    #loop for go through all given post likes
    for i in range(len(l)):
        
        #condition fot viral 
        if l[i]>1000:
            likes[l[i]] =  "viral post"
        else:
            likes[l[i]]= "Normal engagement"
            
    return likes

# list of likes of various post of user
post = [1500,120,200,2045,22100,544]

# call funciton
result = viral(post)

# print information from result 
for i,j in result.items():
    print("total likes: ", i)
    print("post status: ", j)
    print("---------------------------------")
```

# Problem 2: Healthcare – Medicine Stock Alert System


```python
# a function that check stock is sufficient or not for different listed medicine as dictionary
def medicine(lis):

    # dictionary that contain medicine name and stock as key and status as value
    st ={}
    
    #iterate through all medicine in given dictionary 
    for i,j in lis.items():
        #condition for check stock is sufficient or not
        if j >= 10:
            st[f"{i}({j})"] = "stock sufficient"
        else:
            st[f"{i}({j})"] = " low stock alert"
            
    return st

# dictionary that contain available medicine name and stock as well in dictionary format
stock = {
    "paracetamol" :52,
    "savacof":4,
    "flemiclv":15,
    "asprim" :5
}

# call function and store in result
result = medicine(stock)

#print information
for i,j in result.items():
    print("medicine stock: ", i)
    print("status: ", j)
    print("---------------------------------")
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# function that check rainfall is adequate or not
def agriculture(d):
    
    # sum of data values
    s = 0 
    for i in d:
        s += i

    #average of data
    avg = s/len(d)

    #let suppose required level for adequate rainfall is minimum 60 
    req_level = 60

    #condition check
    if avg>=req_level:
        result = "Adequate Rainfall"
    else:
        result = "Inadequate Rainfall"

    return avg, result

# rainfall data 
rainfall_data = [100,20,120,45,75,63]

# call function and return two values avg rainfall and status in form of tuple
result = agriculture(rainfall_data)

#print results
print("Average Rainfall: " , result[0])
print("Rainfall Status: " , result[1] )
```

# Problem 4: Social Media – Duplicate Account Detection

```python
# function that check username is already exist or not (means dupliacte)
#take username list and username that wants to check
def duplicate(username_list, username):

    #find username is found or not in list
    if username in username_list:
        flag = "Yes"
    else:
        flag = "No"

    #return as string with flag value
    return f"Duplicate Account Found : {flag}"

#username list in system
users  = ["ash","misti","tom15","user989","opo785"]

# user input 
user_input = input("enter username to check availability : ")

# calling function
result = duplicate(users,user_input)
print(result)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python
#funtion filter the premium crop price from data of different crop prices
def premium(price):

    #new list that contains premium crop prices only
    premium_crop = []

    #access and check every crop price in list
    for i in price:
        if i >=2000:
            premium_crop.append(i)

    #return list of premium crop  price list
    return f"Premium Crops: {premium_crop}"

# price of various crop in list
crop_prices = [1500,2300,2500,3200,1600,1800,890]

#calling function and print result
result =premium(crop_prices)
print(result)
```

# Problem 8: Banking – Daily Transaction Limit Checker

```python
# this function check transaction is accept or reject according to user daily limit
def transaction(amount):

    # daily limit of user is 50000
    daily_limit = 50000

    #check transaction amount is less than daily limit or not and set flag
    if amount <= daily_limit:
        flag = "Accepted"
    else:
        flag = "Rejected"
        
    return f"Transaction Status: {flag}"
   
# handle wrong input
try:
    amt = int(input("Transaction amount: "))
except:
    print("enter number of error as integer")

#call function
result = transaction(amt)
print(result)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python
#function check student eligibility for exam on the basis of attendance
def eligibility(attendance):

    #sum of present 
    present = 0
    for i in attendance:
        if i == "p":
            present += 1
            
    #attendance percentage
    attendance_percentage = (present/len(attendance))*100

    #check eligibility
    if attendance_percentage >= 75:
        flag = "Eligible"
    else:
        flag = "Not Eligible"
        
    return f"Attendance Percentage: {attendance_percentage:.2f} \nExam Eligibility: {flag}"

#student attendance in list
attendance_data = ["p","p","a","a","p","p","a","p","p","a","a","p","p","a"]

# call function and print result
result =eligibility(attendance_data)
print(result)
```
