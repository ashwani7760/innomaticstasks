## user login check

```python
username = "admin"
password = "1234"

#take username as input
user = input("enter username :")
#take password as input
passw = input("enter password : ")

if username == user and password== passw:
    print("login successful")
else:
    print("invalid credential")
```

## Pass fail analyer

```python
# marks of different student in list marks
marks = [45, 78, 90, 33, 60]
pass_student, failed_student = 0,0
# acess each student marks using loop
for i in marks:
    if i >= 50:
        pass_student += 1
    else:
        failed_student += 1
print("number of student pass :",pass_student)
print("number of student fail :",failed_student)
```

##  simple data cleaner

```python
# list of names
names = [" Alice ", "bob", " CHARLIE "]
for i in range(len(names)):
    names[i]= names[i].replace(" ","").lower() #replace space by empty string and convert into lowercase
print(names)
```

## Message Length Analyzer

## Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for i in logs:
    if i.lower() == "error":
        count += 1
print("number of error occur  : " , count)
```
