```python
class light:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF" #Default
    def turn_on(self):
        self.status = "ON"
    def turn_off(self):
        self.status = "OFF"
    def display_status(self):
        print(f"{self.light_name} is {self.status}")
light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")
l = light(light_name)
if action.upper() == "ON":
    l.turn_on()
elif action.upper() == "OFF":
    l.turn_off()
else:
    print("Invalid Action!")
l.display_status()
```

```python
class employee :
    def __init__(self,name,eid,department):
        self.name = name
        self.eid = eid
        self.department = department
        
    def display(self):
        print("\nEmployee ID Card")
        print("Name: ", self.name)
        print("ID: ", self.eid)
        print("Department: ", self.department)

ename = input("Employee Name: ")
eid   = input("Employee ID: ")
dept = input("Department")

employ = employee(ename,eid,dept)
employ.display()
        
```

```python
class mobile:
    def __init__(self,name,phone):
        self.name = name 
        self.phone  =phone

    def display(self):
        print("\nContact Saved")
        print("Name: ", self.name)
        print("Phone: ", self.phone)

name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

contact = mobile(name, phone)
contact.display()
```

```python
class tag:
    def __init__(self,name,price):
        self.name = name
        self.price=price
    def format(self):
        print("Product: ",self.name)
        print("Price: ",self.price)
prod= input("Product Name: ")
pric = input("Price: ")

product = tag(prod,pric)
product.format()
```

```python
class rating:
    def __init__(self,name,rating):
        self.name = name
        if rating<=5:
            self.rating = rating
        else:
            print("enter rating out of 5")
    def details(self):
        print("\n")
        print("Movie: ", self.name)
        print("Rating: ", self.rating,"/5")

movie_name  = input("Movie Name: ")
rate = float(input("Rating: "))

movie = rating(movie_name,rate)
movie.details()
```
