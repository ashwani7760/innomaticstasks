# problem 1 : unique word in sentence

```python
#take sentence as input
sentence  = input("enter sentence : ")
# split the sentence by space
words = sentence.split()
#convert words list into sentence
unique = set(words)
#get length of set that contain unique words
print("number of unique word : ", len(unique))
```

# problem 2 : highest salary from employee data

```python
#employee data in dictionary format
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
#this variable contain highest salary for start compare we set to 0
highest_salary = 0
#this variable contain name 
names = None
#access each name of employee data
for name in employees:
    #compare highest salary variable with employee salary that select by loop
    if employees[name]>highest_salary:
        highest_salary = employees[name]
        names = name 
print(f"highest salary : {names}- {highest_salary}")
```

# Problem 3: Find Maximum and Minimum Values

```python
#list of value
numbers = [45, 22, 89, 10, 66]
# copy of numbers list
numb = numbers.copy()
# sort the list 
numb.sort()
# mini contain minimum value
mini = numb[0]
#maxi contain maximum value
maxi  = numb[-1]
print("list : ", numbers)
print("min :", mini)
print("max : ", maxi)
```

# Problem 4: Count Products Above a Price Threshold


```python
#list of item price
prices = [450, 1200, 899, 1500, 300]
# count initial 0
count = 0
# iterate price in list
for i in prices:
    # compare price with 1000
    if i > 1000:
        count += 1
#print count
print("product above 1000  : " , count)
```

# Problem 5: Calculate Attendance Percentage

```python
# attendance in list
attendance = ["P", "P", "A", "P", "P"]
# present initial 0
present = 0
# access attendance using loop
for p in attendance:
    if p == "P":
        present  += 1
# total days 
total_days =  len(attendance)
percentage = (present/total_days)*100
print("Attendance Percentage: ", percentage)
```

# Problem 7: Count Character Frequency


```python
#take string as input
string  = input("enter string  : ")
#empty string
freq = {}
#access each character in string
for s in string:
    # it sets s(item of dict) and its value as0 if occur again then +1 the value 
    freq[s] = freq.get(s,0)+1
print(freq)
```

# Problem 9: Check if a Key Exists in a Dictionary

```python
# employee data in dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
#make list of employee name
names = list(employees.keys())
#convert all listed name in lower case
for i in range(len(names)):
    names[i]= names[i].lower()
#take name as input 
emp = input("enter employee name : ").lower()
# check name in list
if emp in names:
    print("Employee exists")
else:
    print("Not exists")
```

# Problem 10: Calculate Average Marks

```python
#list of obtained marks
marks = [20,5,24,15,19]
#list of max marks 
total = [25,25,25,25,25]
# sum initial0
sum_ob,sum_max = 0,0
#loop values 
for i in marks:
    sum_ob += i
for j in total:
    sum_max += j
#average 
avg = (sum_ob/sum_max)*100
print("average marks : ",avg)
```
