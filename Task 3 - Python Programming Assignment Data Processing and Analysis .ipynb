```python
#search query enter by user
query = "Buy mobile phone buy phone online"
#covert query into lower case
lquery = query.lower()
#split query
words = lquery.split()
#keywords
keywords = {}
key={}
for w in words:
    # it sets w(words) and its value as0 if occur again then +1 the value 
    keywords[w] = keywords.get(w,0)+1
    if keywords[w]>1:
        key[w] = keywords.get(w)
print(key)
```

```python
# i solve this problem as multiple different domain is available not only given 2 this code work for all domain
# list of emails
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]

# total number of emails
numofemp = len(emails)

#seperate domain
userdomain=[]
for e in emails:
    userdomain.append(e.split("@")[1])

#dictionary contain domainname and domain frequency
domains = {}
for u in userdomain:
    domains[u] = domains.get(u,0)+1


percent = {}
for dom in domains.keys():
    percent[dom]= f"{(domains[dom]/numofemp)*100}%"

for i,j in percent.items():
    print(f"{i} : {j}")
```
