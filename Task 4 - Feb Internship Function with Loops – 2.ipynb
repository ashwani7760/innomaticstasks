# Problem 1: Mobile Recharge Validation System

```python
# function for recharge validation
def recharge(amount):
    #list of valid plan
    valid_plans = [199, 299, 399, 599]
    #check given conditions
    if amount < 50:
        return "recharge amount should not be less than 50"
    elif amount in valid_plans:
        return True
    else:
        return "invalid amount : valid plans is 199, 299, 399, 599"

# for use while loop
plan = False

#loop run until plan = True
while plan == False:
    amt= int(input("enter recharge amount: "))
    if recharge(amt) == True:
        print("valid plan for recharge")
        break
    else:
        print(recharge(amt))
```

# Problem 2: Inventory Reorder Alert System

```python
# unction for view status of stock of different product maintain in the dictionary
def viewstock(products):
    # dict that contain status
    view = {}
    # loop item 
    for i,j in products.items():
        if j < 15:
             view[i] = f"reorder  product "
        else:
            view[i] =  f"stock ok product"
    return view

#product and stock stored in dictionary
product = {
    "biscuit":50,
    "oil":9,
    "milk" : 12,
    "fruit_combo":19,
    "juice":19,
}

# callfunctiom
print(viewstock(product))
```

# Problem 3: Student Result Processing System

```python
# function
def result(marks):
    if type(marks)== list:
        sum = 0
        # find sum usinf loop
        for i in marks:
            sum += i
        #calculate avg marks
        avg = sum/len(marks)
        # check pass or fail
        if avg >=50:
            return  "Pass"
        else:
            return "fail"
    else:
        return "enter list as input for fuction"
# list of marks
marks = [50,60,20,31,15]

print(result(marks))
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
#funtion for fare calculator take km and peak hr yes or no as input
def fare(km,peak):
    #given rate
    base_fare = 50
    km_fare =12
    if peak.lower()== "y":
        fare  = (base_fare + (km_fare* km))*1.25
    elif peak.lower() == "n":
        fare = base_fare + (km_fare * km)
    return fare

retry = False
# loop
while retry == False:
    try:
        #input by user
        km =int(input("enter km : "))
        peak = input("peak hour(y,n) : ")
        print("fare: ",fare(km,peak))
        print("----------------------------------")
        #if customer wants to retry 
        retry = input("retry(y,n) : ")
        print("----------------------------------")
        if retry.lower() == "n" :
            retry = True
        else:
            retry = False
    except:
        print("enter valid input")
```

# Problem 5: Employee Attendance Eligibility Checker

```python
#function
def eligibility(attendance):
    #count of present "P" in attendance
    present = 0
    for i in attendance:
        if i == "P":
            present +=1
    #percentage of present 
    percent = (present/len(attendance)) * 100
    #check condition for eligible
    if percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# list of attendance
attendance = ["A","P","P","P","P","P","P","A","P","P","P","P","A","A","A"]

# call function
print(eligibility(attendance))
```

# Problem 6: Password Strength Checker

```python
#function 
def strength(password):
    #let all condition is false
    length,inte,special = False,False,False
    #change flag(length, inte, special) as true if condition met
    if len(password) > 8 :
        length = True
    for i in password:
        if i in "0123456789" :
            inte = True
    for i in password:
        if i in "@#$":
            special = True

    # return based on password
    if length is False:
        return "length atleast of 8"
    elif inte is False:
        return "password Must contain at least one digit"
    elif special is False:
        return "password Must contain at least one special character (@#$)"
    else:
        return "Strong password"
        
pswrd = input("enter password: ")
print(strength(pswrd))
```
