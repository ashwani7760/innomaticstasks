# Problem Statement 1:Smart Parking Lot Management System

```python
def parking(capacity, logs):
    #  current parked variable
    current_parked = 0
    # store max usgae of parking
    peak_usage = 0

    # calculate parked vehicle
    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1
        
        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Determine parking status
    if current_parked > capacity:
        status = "Alert: Parking Exceeded Capacity!"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

print(parking(capacity, logs))
```

# Problem Statement 2:Online Food Delivery Time Estimator

```python
def delivery_time(distance, traffic, weather):
    # Base speed assumption (km per hour)
    speed = 40  
    
    # Calculate base time in minutes
    time = (distance / speed) * 60
    
    # Traffic adjustments
    if traffic.lower() == "low":
        traffic_delay = 0
    elif traffic.lower() == "medium":
        traffic_delay = 10
    elif traffic.lower() == "high":
        traffic_delay = 20
    else:
        traffic_delay = 5  # default delay
    
    # Weather adjustments
    if weather.lower() == "clear":
        weather_delay = 0
    elif weather.lower() == "rainy":
        weather_delay = 15
    elif weather.lower() == "stormy":
        weather_delay = 25
    else:
        weather_delay = 5  # default delay
    
    # Final ETA
    total_time = time + traffic_delay + weather_delay
    
    return round(total_time)


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

eta = delivery_time(distance, traffic, weather)
print(f"Estimated Delivery Time: {eta} minutes")
```

 # Problem Statement 3:Movie Theatre Seat Occupancy Analyzer

```python
def theatre(total_seats,booked_seats):
    # total booked seats
    booked_ticket = len(booked_seats)
    # occupancy occupied percentage
    occupancy = (booked_ticket / total_seats) * 100
    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Filling Fast"
    else:
        status = "Available"
    return f"Occupancy: {occupancy}% \nShow status: {status}"
    
# for book ticket for problem at once
l = []
for i in range(90):
    l.append(1)
#max seats or ticket will be 
total = 150
the = theatre(total,l)
print(the)
```

#  Problem Statement 4:Cloud Server Load Classification System

```python
def server(readings):
    # calculate average of cpu readings
    sumofread = 0
    for reading in readings:
        sumofread += reading
    avg_reading = sumofread / len(readings)

    # determining status based on avg cpu usage 
    if avg_reading<= 50 :
        status = "Normal"
    elif avg_reading <80:
        status = "Warning"
    else:
        status = "Critical"
    return f"Average CPU Load: {avg_reading}% \nServer Status: {status}"

#sample input
cpu_readings= [45, 60, 70, 85, 90]
# calling function and print return
result = server(cpu_readings)
print(result)
```

# Problem Statement 6: Online Event Registration Capacity Controller

```python
def event(event_capacity, reg):
    # max registeration that include waitlist 
    max_reg = 110
    waitlist = 0
    status = "Available"
    #determine confirmed and waitlist and status  of reggistration
    if reg <=  event_capacity:
        confirm = reg
    elif reg < max_reg:
        confirm =  event_capacity
        waitlist = reg- confirm
    else:
        confirm =  event_capacity
        waitlist = max_reg- event_capacity
        status = "Closed"

    return f"Confirm Registration: {confirm} \nWaitlisted User: {waitlist} \nRegistration status: {status}"

#111 user registered
registered = 111
#100 user can join the event
max_joiner = 100
event1 = event(max_joiner,registered)
print(event1)
```
